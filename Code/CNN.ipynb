```python
def copy_images(source_folder, destination_folder, num_images):
    # Ensure the destination folder exists
    os.makedirs(destination_folder, exist_ok=True)

    image_files = sorted(os.listdir(source_folder))[:num_images]  # Select only the first `num_images`

    # Iterate through all files in the source folder
    for file_name in image_files:
        source_path = os.path.join(source_folder, file_name)

        # Only proceed if it's a file (skip subdirectories)
        if os.path.isfile(source_path):
            destination_path = os.path.join(destination_folder, file_name)

            # Copy the file without checking for conflicts
            shutil.copy(source_path, destination_path)
```

# Training CNN for Different Training Sets

```python
def get_dense_factor(filter_config):
    if filter_config == [16, 32, 64]:
        return 256
    elif filter_config == [16, 32, 64, 128]:
        return 512
    elif filter_config == [16, 32, 64, 128, 256]:
        return 1024
    else:
        raise ValueError('Unknown filter configuration.')
```

```python
def create_model(filter_config, input_shape=(128, 128,1)):
    model = Sequential()
    model.add(tf.keras.Input(shape=input_shape))
    
    for num_filters in filter_config:
        model.add(Conv2D(num_filters, (3, 3), activation='relu', padding='same'))
        model.add(MaxPooling2D(pool_size=(2, 2)))
      
    model.add(Flatten())
    model.add(Dense(get_dense_factor(filter_config), activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(26, activation='softmax'))
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python
def create_datagen_without_DA():
    
    return ImageDataGenerator(
        rescale=1.0 / 255,
        validation_split=0.2
    )
```

```python
# Define gaussian blur as an option for data augmentation, since it is not contained in keras.
def gaussian_blur(image, kernel_size):
    if kernel_size:
        image = cv2.GaussianBlur(image, kernel_size, 0)
    
    if len(image.shape) == 2:  # Add channel dimension if missing
        image = np.expand_dims(image, axis=-1)
        
    return image
```

```python
def create_datagen_with_DA(skeleton=True): 
    # define rotation range
    if skeleton:
        rot = 10
        shear = 0.0
        preprocessing_function = None
    else:
        rot = 15
        shear = 0.2
        preprocessing_function = (lambda image: gaussian_blur(image, (5, 5)))
    
    return ImageDataGenerator(
        rescale=1.0 / 255,
        rotation_range = rot,
        shear_range = shear,
        preprocessing_function = preprocessing_function,
        validation_split=0.2
    )
```

```python
# Define callbacks
early_stopping = EarlyStopping(
    monitor='val_loss',            # Monitor validation loss
    patience=5,                    # Stop if no improvement for 5 epochs
    restore_best_weights=True      # Restore weights of the best model
)

model_checkpoint = ModelCheckpoint(
    'best_model.keras',            # Save the best model as .keras file
    save_best_only=True,           # Only save if current model is better
    monitor='val_loss'             # Monitor validation loss for "best" model
)
```

```python
def determine_situation(train_directory_, with_DA):
    splitted_td = train_directory_.split('/')
    
    scenario = splitted_td[-2]
    train_set = splitted_td[-1]
    da = 'DA_True' if with_DA else 'DA_False'
    
    return scenario + '//' + train_set + '//' + f'{da}'
```

```python
def train_and_evaluate(train_directory, test_directory, with_DA=True, skeleton=True):
    result = {}
    situation = determine_situation(train_directory, with_DA)
    
    try:
        print(f"Training with dataset: {situation}")

        if with_DA:
            datagen = create_datagen_with_DA(skeleton=skeleton)
        else:
            datagen = create_datagen_without_DA()

        train_generator = datagen.flow_from_directory(
            train_directory,
            target_size=(128, 128),
            color_mode='grayscale',
            batch_size=32,
            class_mode='categorical',
            subset='training',
        )

        valid_generator = datagen.flow_from_directory(
            train_directory,
            target_size=(128, 128),
            color_mode='grayscale',
            batch_size=32,
            class_mode='categorical',
            subset='validation',
        )

        test_datagen = ImageDataGenerator(rescale=1.0 / 255)
        test_generator = test_datagen.flow_from_directory(
            test_directory, 
            target_size=(128, 128),
            color_mode='grayscale',
            batch_size=32,
            class_mode='categorical',
        )

        # Recalculate steps
        steps_per_epoch = len(train_generator)
        validation_steps = len(valid_generator)
        test_steps = len(test_generator)

        print(f"Steps per epoch: {steps_per_epoch}, Validation steps: {validation_steps}, Test steps: {test_steps}")

        if not with_DA and not skeleton:
            model = create_model([16, 32, 64, 128, 256])
        elif not with_DA and skeleton:
            model = create_model([16, 32, 64, 128])
        elif with_DA and not skeleton:
            model = create_model([16, 32, 64, 128])
        else:
            model = create_model([16, 32, 64, 128, 256])

        history = model.fit(
            train_generator,
            validation_data=valid_generator,
            epochs=10,
            callbacks=[early_stopping, model_checkpoint],
            steps_per_epoch=steps_per_epoch,
            validation_steps=validation_steps          
        ) 

        loss, accuracy = model.evaluate(test_generator, steps=test_steps, verbose=0)
        print(f"Test Loss: {loss}, Test Accuracy: {accuracy}")

        result = {
            "situation": situation,
            "loss": loss,
            "accuracy": accuracy,
            "history": history
        }

    except Exception as e:
        print(f"Error for the situation: {situation}")
        print(e)

    finally:
        if 'model' in locals():
            del model
        gc.collect()
        
    return result
```

## Visualization

```python
def show_results(results):
    # Process the results to organize by scenario and data augmentation
    scenario_1_false = []
    scenario_1_true = []
    scenario_2_false = []
    scenario_2_true = []
    
    for entry in results:
        situation = entry['situation']
        loss = round(entry['loss'], 3)
        accuracy = round(entry['accuracy'], 3)
        
        if 'Scenario_1' in situation and 'DA_False' in situation:
            scenario_1_false.append(accuracy)
        elif 'Scenario_1' in situation and 'DA_True' in situation:
            scenario_1_true.append(accuracy)
        elif 'Scenario_2' in situation and 'DA_False' in situation:
            scenario_2_false.append(accuracy)
        elif 'Scenario_2' in situation and 'DA_True' in situation:
            scenario_2_true.append(accuracy)
    
    # Standard x-axis labels
    config_labels = ['original', 'original+T', 'original+2T', 'original+3T', 'original+4T', 'original+5T']
    
    # Plotting all data in a single graph
    plt.figure(figsize=(12, 8))
    plt.plot(config_labels, scenario_1_false, marker='o', label='Scenario 1, Baseline 1', linestyle='--')
    plt.plot(config_labels, scenario_1_true, marker='o', label='Scenario 1, Baseline 2', linestyle='-')
    plt.plot(config_labels, scenario_2_false, marker='o', label='Scenario 2, Baseline 1', linestyle='--')
    plt.plot(config_labels, scenario_2_true, marker='o', label='Scenario 2, Baseline 2', linestyle='-')
    
    # Customize the plot
    plt.title('Accuracy for Training Configurations (All Scenarios)', fontsize=14)
    plt.xlabel('Training Configuration', fontsize=12)
    plt.ylabel('Accuracy', fontsize=12)
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    
    # Show the plot
    plt.show()
```
